# Task 2

Exp($\lambda$); медиана; $\lambda=1$.

Воспользуемся теоремой об асимптотическом поведении среднего члена вариационного ряда:

$$ \sqrt n \frac{(X_{(n \alpha)} - q_{\alpha})}{\sqrt{\alpha (1-\alpha)}} p(q_{\alpha}) \rightarrow^d U \sim \mathcal{N} (0, 1) $$

$$ \sqrt n \frac{(X_{\frac{n}{2}} - q_{\frac{1}{2}})}{\sqrt{\frac{1}{2} (1-\frac{1}{2})}} p(q_{\frac{1}{2}}) \rightarrow^d U \sim \mathcal{N} (0, 1) $$

$$ 2 \sqrt n (M - q_{\frac{1}{2}}) p(q_{\frac{1}{2}}) \rightarrow^d U \sim \mathcal{N} (0, 1) $$, где $M$ это медиана.

Квантиль экспоненциального распределения:

$q_{\frac{1}{2}} = \frac{- \ln (1 - \frac{1}{2})}{\lambda} = \frac{\ln 2}{\lambda}$

$p(q_{\alpha}) = \lambda e^{-\lambda\frac{\ln 2}{\lambda}}=\frac{\lambda}{2}$

$$ 2 \sqrt n (M - \frac{\ln 2}{\lambda}) \frac{\lambda}{2} \rightarrow^d U \sim \mathcal{N} (0, 1) $$

$$ \lambda\sqrt n (M - \frac{\ln 2}{\lambda}) \rightarrow^d U \sim \mathcal{N} (0, 1) $$

Тогда возьмем квантиль нормального распределения:

$$P[-q_{1-\frac{\alpha}{2}} \leq \lambda\sqrt n (M - \frac{\ln 2}{\lambda}) \leq q_{1-\frac{\alpha}{2}}] = 1 - \alpha$$

Тогда 

$$P[-q_{1-\frac{\alpha}{2}} \leq \lambda\sqrt n M - \sqrt n \ln 2 \leq q_{1-\frac{\alpha}{2}}] = 1 - \alpha$$
$$P[-q_{1-\frac{\alpha}{2}} + \sqrt n \ln 2 \leq \lambda\sqrt n M \leq q_{1-\frac{\alpha}{2}}+\sqrt n \ln 2] = 1 - \alpha$$
$$P( \frac{-q_{1-\frac{\alpha}{2}} + \sqrt n \ln 2}{\sqrt n M} \leq \lambda \leq \frac{q_{1-\frac{\alpha}{2}}+\sqrt n \ln 2}{\sqrt n M} ) = 1 - \alpha$$
Теоретическая медиана это $\frac{\ln2}{\lambda}$, оценим её
$$P( \frac{\sqrt n M \ln2}{-q_{1-\frac{\alpha}{2}} + \sqrt n \ln 2} \geq \frac{\ln2}{\lambda} \geq \frac{\sqrt n M \ln2 }{q_{1-\frac{\alpha}{2}}+\sqrt n \ln 2} ) = 1 - \alpha$$
$$P( \frac{\sqrt n M \ln2 }{q_{1-\frac{\alpha}{2}}+\sqrt n \ln 2} \leq \frac{\ln2}{\lambda} \leq \frac{\sqrt n M \ln2}{-q_{1-\frac{\alpha}{2}} + \sqrt n \ln 2} ) = 1 - \alpha$$

Тут $M$ --- медиана, $n$ --- размер нашей выборки.

```python
import scipy.stats as stats
import math

lam = 1
alpha = 0.05
N = 25
M = 25

q_r = stats.norm.ppf(1 - alpha / 2)

rndX = stats.expon(lam)
```

```python
def gen_sample(size):
    return [rndX.rvs() for _ in range(size)]
```

```python
def med(xs):
    return sorted(xs)[len(xs) // 2]


def calc_interval(n, m):
    left = math.sqrt(n) * m * math.log(2) / (q_r + math.log(2) * math.sqrt(n))
    right = math.sqrt(n) * m * math.log(2) / (-q_r + math.log(2) * math.sqrt(n))
    return left, right
```

```python
def check_interval(amount):
    res = 0
    for _ in range(amount):
        m = med(gen_sample(N))
        left, right = calc_interval(N, m)
        # print(left, right)
        if left <= math.log(2) / lam <= right:
            res += 1
    print(f"{res}/{amount} \t\t {res / amount * 100:.2f}%")
```

```python
check_interval(1000)
```

```python
check_interval(50000)
```
